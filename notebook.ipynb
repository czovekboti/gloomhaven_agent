## Run to evaluate agent

```python
import sys
import os
from pathlib import Path

project_root = str(Path.cwd())
if project_root not in sys.path:
    sys.path.append(project_root)

from tests.evaluate_agent import evaluate_agent
import pandas as pd

report = evaluate_agent()

print("\n--- Quick Summary ---")
print(f"Evaluated: {report['total_evaluated']}")
print(f"Exact Correct Accuracy: {report['exact_correct_accuracy_percentage']:.2f}%")

if report.get("failed_evaluations"):
    print("\n--- Failed Evaluations ---")
    df_failed = pd.DataFrame(report["failed_evaluations"])
    display(df_failed[[
        'question',
        'expected_correct',
        'agent_prediction',
        'agent_explanation'
    ]])

successful_evals = report.get('successful_evaluations', [])
if successful_evals:
    print("\n--- SUCCESSFUL EVALUATIONS ---")
    df_success = pd.DataFrame(successful_evals)
    df_display = df_success[[
        'question',
        'agent_explanation',
        'explanation_review'
    ]].rename(columns={
        'question': 'Question',
        'agent_explanation': 'Agent Explanation',
        'explanation_review': 'Judge Review'
    })
    styled_df = df_display.style.set_properties(**{
        'text-align': 'left',
        'white-space': 'pre-wrap',
        'vertical-align': 'top'
    }).set_table_styles([
        {'selector': 'th', 'props': [('text-align', 'left'), ('font-weight', 'bold')]}
    ])
    display(styled_df)
```

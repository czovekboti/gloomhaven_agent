```python
import sys
import os
from pathlib import Path
project_root = str(Path.cwd())
if project_root not in sys.path:
    sys.path.append(project_root)

from tests.evaluate_agent import evaluate_agent

report = evaluate_agent()
import pandas as pd

print("\n--- Quick Summary ---")
print(f"Evaluated: {report['total_evaluated']}")
print(f"Exact Correct Accuracy: {report['exact_correct_accuracy_percentage']:.2f}%")
print(f"LLM Judge Accuracy: {report['llm_accuracy_percentage']:.2f}%")
if report.get("failed_evaluations"):
    print("\n--- Failed Evaluations ---")
    df_failed = pd.DataFrame(report["failed_evaluations"])
    display(df_failed[[
        'question', 
        'expected_correct', 
        'agent_prediction', 
        'agent_explanation'
    ]])
successful_evals = report.get('successful_evaluations', []),
if successful_evals:
    print("\n--- Successful Evaluations ---")
    df_successful = pd.DataFrame(successful_evals)
    cols = df_successful.columns.tolist()
    priority_cols = ['question', 'expected_correct', 'agent_prediction', 'expected_category', 'agent_category']
    reordered_cols = [c for c in priority_cols if c in cols] + [c for c in cols if c not in priority_cols]
    display(df_successful[reordered_cols])
```
